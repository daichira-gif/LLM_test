```python
import pandas as pd

# データを読み込む
url = "https://www.mhlw.go.jp/content/12400000/001122669.xlsx"
df = pd.read_excel(url)

# セル結合を解除
df = df.fillna(method='ffill')

# 先頭行と４番目の列をインデックスとする
df.set_index(df.columns[0], inplace=True)
df.index.name = None
df.columns = df.iloc[2]
df = df[4:]


```

```python
# 2列目以降のデータを全てint型に変換し、エラーが返されたらnullにする
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# 1列目のインデックスを「BMI検査値階層(kg/㎡)」に書き換える
df.columns.values[0] = 'BMI検査値階層(kg/㎡)'

# 2列目から10列目までのインデックスの頭に「男_」を追加
df.columns.values[1:9] = '男_' + df.columns[1:9]

# 11列目から18列目までのインデックスの頭に「女_」を追加
df.columns.values[9:17] = '女_' + df.columns[9:17]


# データフレームを'BMI'という名前のExcelファイルとして保存
df.to_excel('BMI.xlsx')

df
```
